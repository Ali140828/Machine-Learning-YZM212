## Characteristic Polynomial and Eigenvalues

To find the eigenvalues of a matrix \( A \), we use the characteristic equation:

$$
\det(A - \lambda I) = 0
$$

**Where:**

- $A$ : Given matrix  
- $\lambda$ : Eigenvalues  
- $I$ : Identity matrix  
- $\det$ : Determinant  

For a $2 \times 2$ matrix:

$$
A =
\begin{bmatrix}
a & b \\
c & d
\end{bmatrix}
$$

The identity matrix is:

$$
I =
\begin{bmatrix}
1 & 0 \\
0 & 1
\end{bmatrix}
$$

Then:

$$
A - \lambda I =
\begin{bmatrix}
a - \lambda & b \\
c & d - \lambda
\end{bmatrix}
$$

The determinant is:

$$
\det(A - \lambda I) = (a - \lambda)(d - \lambda) - bc
$$

Expanding:

$$
(a - \lambda)(d - \lambda) = ad - a\lambda - d\lambda + \lambda^2
$$

So the characteristic polynomial becomes:

$$
\lambda^2 - (a + d)\lambda + (ad - bc) = 0
$$

Using:

- Trace:
$$
\text{trace}(A) = a + d
$$

- Determinant:
$$
\det(A) = ad - bc
$$

We get:

$$
\lambda^2 - \text{trace}(A)\lambda + \det(A) = 0
$$

Solving this equation gives the eigenvalues of the matrix.

## Kodda (eig olmadan)

```python
import numpy as np

# Define matrix
A = np.array([[2, 1],
              [1, 2]])


trace = np.trace(A)
det = np.linalg.det(A)

print("Trace:", trace)
print("Determinant:", det)
```

$
\lambda^2 - \text{trace}(A)\lambda + \det(A) = 0
$

```python
coefficients = [1, -trace, det]

eigenvalues = np.roots(coefficients)

print("Eigenvalues (manual):")
print(eigenvalues)
```

## Kodda (eig ile)

```python
import numpy as np

A = np.array([[2, 1],
              [1, 2]])

print(A,end="\n\n")

values, vectors = np.linalg.eig(A)

print("Eigenvalues:")
print(values)

print("Eigenvectors:")
print(vectors)
```
