# Eigenvalues and Eigenvectors

Bu notebook'ta özdeğerler (eigenvalues) ve özvektörler (eigenvectors) hesaplanacaktır.

```python
import numpy as np
```

```python
A = np.array([[2, 1],
              [1, 2]])

A
```

```python
values, vectors = np.linalg.eig(A)

print("Eigenvalues:")
print(values)

print("Eigenvectors:")
print(vectors)
```

## Sonuçların Yorumu

Bu çalışmada elde edilen özdeğerler 3 ve 1'dir. Bu değerler, matrisin belirli yönlerde veri üzerinde nasıl bir dönüşüm yaptığını gösterir.

Özvektörler ise dönüşüm sırasında yönü değişmeyen vektörlerdir. Sadece büyüklükleri özdeğer kadar ölçeklenir.

Bu nedenle özvektörler, veri içindeki önemli yönleri temsil eder.
